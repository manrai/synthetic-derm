```python
import os
```

```python
def print_directory_tree(directory, level=0):
    try:
        # List all items in the directory
        items = os.listdir(directory)
    except PermissionError:
        # Skip directories that cannot be accessed due to permission errors
        print(" " * (level * 4) + f"Permission Denied: {directory}")
        return

    # Separate files and directories
    file_count = sum(1 for item in items if os.path.isfile(os.path.join(directory, item)))
    subdirectories = [item for item in items if os.path.isdir(os.path.join(directory, item))]

    # Print directory name with counts of files and subdirectories
    print(" " * (level * 4) + f"{os.path.basename(directory)} - {len(subdirectories)} subdirectories, {file_count} files")

    # Recurse into each subdirectory
    for subdirectory in subdirectories:
        subdirectory_path = os.path.join(directory, subdirectory)
        print_directory_tree(subdirectory_path, level + 1)
```

```python
def count_files_in_directory(directory):
    try:
        # List all items in the directory
        items = os.listdir(directory)
    except PermissionError:
        print(f"Permission Denied: {directory}")
        return None

    # Count files only (ignore subdirectories)
    file_count = sum(1 for item in items if os.path.isfile(os.path.join(directory, item)))

    print(f"{directory} contains {file_count} files.")
```
